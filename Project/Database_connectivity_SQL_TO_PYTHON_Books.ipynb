```python
import pyodbc

try:
    conn = pyodbc.connect(
        r"Driver={ODBC Driver 17 for SQL Server};"
        r"Server=LAPTOP-DTL19BIG\SQLEXPRESS;"
        r"Database=SchoolBooks;"
        r"Trusted_Connection=yes;"
    )

    cursor = conn.cursor()

    print("Welcome to School Book Store")

    grand_total = 0
    items = []

    while True:

        print("\n1. Buy Book")
        print("2. Buy Notebook")

        choice = input("Enter your choice (1 or 2): ")

        # BOOK SECTION
        if choice == "1":

            standard = input("Enter Standard (1st-4th / 5th-8th / 9th-10th): ")

            cursor.execute("SELECT book_id, subject, price FROM Books WHERE standard=?", standard)
            books = cursor.fetchall()

            print("\nAvailable Books:")
            for b in books:
                print("ID:", b.book_id, "|", b.subject, "| Price:", b.price)

            book_id = int(input("Enter Book ID: "))
            quantity = int(input("Enter Quantity: "))

            cursor.execute("SELECT subject, price FROM Books WHERE book_id=?", book_id)
            book = cursor.fetchone()

            if book:
                name = book.subject
                price = book.price
                total = price * quantity

                cursor.execute(
                    "INSERT INTO Orders (item_name, quantity, price, total) VALUES (?, ?, ?, ?)",
                    name, quantity, price, total
                )
                conn.commit()

                items.append((name, quantity, price, total))
                grand_total += total
            else:
                print("Invalid Book ID")

        # NOTEBOOK SECTION
        elif choice == "2":

            cursor.execute("SELECT notebook_id, type, pages, price FROM Notebooks")
            notebooks = cursor.fetchall()

            print("\nAvailable Notebooks:")
            for n in notebooks:
                print("ID:", n.notebook_id, "|", n.type, "-", n.pages, "pages | Price:", n.price)

            nb_id = int(input("Enter Notebook ID: "))
            quantity = int(input("Enter Quantity: "))

            cursor.execute("SELECT type, pages, price FROM Notebooks WHERE notebook_id=?", nb_id)
            nb = cursor.fetchone()

            if nb:
                name = nb.type + " " + str(nb.pages) + " pages"
                price = nb.price
                total = price * quantity

                cursor.execute(
                    "INSERT INTO Orders (item_name, quantity, price, total) VALUES (?, ?, ?, ?)",
                    name, quantity, price, total
                )
                conn.commit()

                items.append((name, quantity, price, total))
                grand_total += total
            else:
                print("Invalid Notebook ID")

        else:
            print("Wrong choice")

        more = input("Do you want to add more items? (yes/no): ")
        if more.lower() == "no":
            break

    
    print("\n====================================================")
    print("                    FINAL BILL                        ")
    print("====================================================")

    print(f"{'Item':30} {'Qty':5} {'Price':10} {'Total':10}")
    print("----------------------------------------------------")

    for name, qty, price, total in items:
        print(f"{name:30} {qty:<5} {price:<10.2f} {total:<10.2f}")

    print("----------------------------------------------------")
    print(f"{'Grand Total':30} {'':5} {'':10} {grand_total:<10.2f}")
    print("====================================================")

    conn.close()

except ValueError:
    print("Please enter correct number values.")

except pyodbc.Error as e:
    print("Database Error:", e)

except Exception as e:
    print("Something went wrong:", e)
```
