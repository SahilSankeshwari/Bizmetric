# Hotel Bill Printing Task

```python
import pyodbc

conn = pyodbc.connect(
    r"Driver={ODBC Driver 17 for SQL Server};"
    r"Server=LAPTOP-DTL19BIG\SQLEXPRESS;"
    r"Database=Menu;"
    r"Trusted_Connection=yes;"
)

cursor = conn.cursor()
cursor.execute("SELECT * FROM Menu")

# records=cursor.fetchall()
# print(records)
for row in cursor:
   print(row)
```

```python
print("Welcome to Nil-Kamal Hotel")
print("\n----------- HOTEL MENU -----------")

cursor.execute("SELECT * FROM Menu")

for row in cursor:
    print(f"ID:{row.id} | Item:{row.item_name} | Price:Rs.{row.price}")

grand_total = 0
items = []

while True:

    item_id = int(input("\nEnter Item ID: "))
    quantity = int(input("Enter Quantity: "))

    cursor.execute("SELECT item_name, price FROM Menu WHERE id=?", item_id)
    item = cursor.fetchone()

    if item is None:
        print("Invalid Item ID")
    else:
        item_name = item.item_name
        price = item.price
        total = price * quantity

        #insert in database
        cursor.execute(
            "INSERT INTO Orders (item_id, quantity, price, total) VALUES (?, ?, ?, ?)",
            item_id, quantity, price, total
        )
        conn.commit() #save in database

        items.append((item_name, quantity, price, total)) # Store for final bill
        grand_total += total

    choice = input("Do you want to order more? (yes/no): ")
    if choice.lower() == "no":
        break


# Printing Bill
print("\n=========== BILL ===========")

for i in items:
    print("Item:", i[0], "| Qty:", i[1], "| Price:", i[2], "| Total:", i[3])

print("----------------------------------")
print("Grand Total:", grand_total)

conn.close()
```
