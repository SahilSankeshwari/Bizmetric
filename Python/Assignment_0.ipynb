#LIST COMPREHENSION

```python
# 1. Generate squares of numbers from 1 to 10.
squares = [x**2 for x in range(1, 11)]
print(squares)
```

```python
# 2. Create a list of even numbers between 1 and 50.
even_numbers = [x for x in range(1, 51) if x % 2 == 0]
print(even_numbers)
```

```python
# 3. Convert all strings in a list to uppercase.
a = input("Enter words separated by space: ").split()
b = [i.upper() for i in a]
print(b)
```

```python
# 4. Create a new list that contains only positive numbers.
a = list(map(int, input("Enter numbers: ").split()))
b = [i for i in a if i > 0]
print(b)
```

```python
# 5. Create list of tuples (num, num^2) for numbers 1 to 5.
a = [(i, i**2) for i in range(1, 6)]
print(a)
```

```python
# 6. Extract all vowels from a string.
a = input("Enter a string: ")
b = [i for i in a if i.lower() in "aeiou"]
print(b)
```

```python
# 7. Flatten a 2D list.
a = [[1,2],[3,4],[5,6]]
b = [j for i in a for j in i]
print(b)
```

```python
# 8. Replace negative numbers with 0.
a = list(map(int, input("Enter numbers: ").split()))
b = [i if i > 0 else 0 for i in a]
print(b)
```

```python
# 9. Create list of lengths of each word.
a = input("Enter words: ").split()
b = [len(i) for i in a]
print(b)
```

```python
# 10. Filter out words starting with 'A' or 'a'.
a = input("Enter words: ").split()
b = [i for i in a if not i.lower().startswith('a')]
print(b)
```

```python
# 11. Generate "even" or "odd" list.
a = list(map(int, input("Enter numbers: ").split()))
b = ["even" if i%2==0 else "odd" for i in a]
print(b)
```

```python
# 12. Numbers divisible by 3 and 5 in range 1–100.
a = [i for i in range(1,101) if i%3==0 and i%5==0]
print(a)
```

```python
# 13. Multiplication table 1–5.
a = [[i*j for j in range(1,6)] for i in range(1,6)]
print(a)
```

```python
# 14. Convert dictionary keys into list.
a = {"x":1,"y":2,"z":3}
b = [i for i in a]
print(b)
```

```python
# 15. Extract digits from a string.
a = input("Enter a string: ")
b = [i for i in a if i.isdigit()]
print(b)
```

```python
# 16. Remove all spaces from a string.
a = input("Enter a string: ")
b = [i for i in a if i != " "]
print("".join(b))
```

```python
# 17. Characters appearing more than once.
a = input("Enter a string: ")
b = [i for i in a if a.count(i) > 1]
print(list(set(b)))
```

```python
# 18. Generate list of all words from sentences.
a = input("Enter sentences separated by comma: ").split(",")
b = [j for i in a for j in i.split()]
print(b)
```

```python
# 19. Create list of unique elements.
a = list(map(int, input("Enter numbers: ").split()))
b = [i for i in set(a)]
print(b)
```

```python
# 20. Generate all pairs (x,y) from two lists.
a = list(map(int, input("Enter first list: ").split()))
b = list(map(int, input("Enter second list: ").split()))
c = [(i,j) for i in a for j in b]
print(c)
```

# LAMBDA FUNCTIONS

```python
from functools import reduce
import math
```

```python
# 1. Add two numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = lambda x,y: x+y
print(c(a,b))
```

```python
# 2. Check if number is even.
a = int(input("Enter number: "))
b = lambda x: x%2==0
print(b(a))
```

```python
# 3. Get last character of string.
a = input("Enter string: ")
b = lambda x: x[-1]
print(b(a))
```

```python
# 4. Square every number in list.
a = list(map(int,input("Enter numbers: ").split()))
b = list(map(lambda x: x**2, a))
print(b)
```

```python
# 5. Get only odd numbers.
a = list(map(int,input("Enter numbers: ").split()))
b = list(filter(lambda x: x%2!=0, a))
print(b)
```

```python
# 6. Sort list of tuples by second value.
a = [(1,3),(2,1),(4,2)]
b = sorted(a, key=lambda x: x[1])
print(b)
```

```python
# 7. Check palindrome.
a = input("Enter string: ")
b = lambda x: x == x[::-1]
print(b(a))
```

```python
# 8. Find maximum of three numbers.
a,b,c = map(int,input("Enter three numbers: ").split())
d = lambda x,y,z: max(x,y,z)
print(d(a,b,c))
```

```python
# 9. Reverse string.
a = input("Enter string: ")
b = lambda x: x[::-1]
print(b(a))
```

```python
# 10. Convert list of strings to integers.
a = input("Enter numbers: ").split()
b = list(map(lambda x:int(x), a))
print(b)
```

```python
# 11. Remove empty strings.
a = input("Enter words separated by space: ").split()
b = list(filter(lambda x: x!="", a))
print(b)
```

```python
# 12. Compute factorial using reduce.
from functools import reduce
a = int(input("Enter number: "))
b = reduce(lambda x,y: x*y, range(1,a+1))
print(b)
```

```python
# 13. Return larger of two numbers.
a,b = map(int,input("Enter two numbers: ").split())
c = lambda x,y: x if x>y else y
print(c(a,b))
```

```python
# 14. Check divisible by 5.
a = int(input("Enter number: "))
b = lambda x: x%5==0
print(b(a))
```

```python
# 15. Add 10 to each element.
a = list(map(int,input("Enter numbers: ").split()))
b = list(map(lambda x: x+10, a))
print(b)
```

```python
# 16. Sort list of dictionaries by age.
a = [{"name":"Sahil","age":25},{"name":"Abhi","age":20}]
b = sorted(a, key=lambda x: x["age"])
print(b)
```

```python
# 17. Check if character is vowel.
a = input("Enter character: ")
b = lambda x: x.lower() in "aeiou"
print(b(a))
```

```python
# 18. Extract words length > 5.
a = input("Enter words: ").split()
b = list(filter(lambda x: len(x)>5, a))
print(b)
```

```python
# 19. Area of circle.
import math
a = float(input("Enter radius: "))
b = lambda x: math.pi*x*x
print(b(a))
```

```python
# 20. Remove duplicates.
a = list(map(int,input("Enter numbers: ").split()))
b = lambda x: list(set(x))
print(b(a))
```

```python
# 21. Product of all numbers.
a = list(map(int,input("Enter numbers: ").split()))
b = reduce(lambda x,y: x*y, a)
print(b)
```

```python
# 22. Absolute value.
a = int(input("Enter number: "))
b = lambda x: abs(x)
print(b(a))
```

```python
# 23. Sort strings by length.
a = input("Enter words: ").split()
b = sorted(a, key=lambda x: len(x))
print(b)
```

```python
# 24. Get only uppercase characters.
a = input("Enter string: ")
b = [i for i in a if i.isupper()]
print(b)
```

```python
# 25. Square if even, cube if odd.
a = int(input("Enter number: "))
b = lambda x: x**2 if x%2==0 else x**3
print(b(a))
```

```python
# 26. Celsius to Fahrenheit.
a = float(input("Enter Celsius: "))
b = lambda x: (x*9/5)+32
print(b(a))
```

```python
# 27. Check anagrams.
a = input("Enter first string: ")
b = input("Enter second string: ")
c = lambda x,y: sorted(x)==sorted(y)
print(c(a,b))
```

```python
# 28. Extract numeric values from mixed list.
a = [1,"a",3.5,"b",7]
b = list(filter(lambda x: isinstance(x,(int,float)), a))
print(b)
```

```python
# 29. Check if any element is negative.
a = list(map(int,input("Enter numbers: ").split()))
b = any(map(lambda x: x<0, a))
print(b)
```

```python
# 30. Generate function that multiplies by n.
a = int(input("Enter multiplier: "))
b = lambda n: lambda x: x*n
c = b(a)
d = int(input("Enter number: "))
print(c(d))
```
