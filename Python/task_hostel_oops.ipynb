Ask user to opt for courses for master degree based on the following
L1 = [“HR”, “Finance”, “Marketing”, “DS”]
Based on above subject there are two different streams. For example- HR is having HR core and HR analytics and Marketing is having core and Marketing analytics. Analytics is the optional subject and having added extra fees. DS is not having analytics.
If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.
If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 .
Transportation charges 13000 per semester. Calculate the total annual cost based on selected service.  
User will enter values as subject, analytics(Y/N), Hostel (Y/N), food(How many months?), Transportation(semester/annual)


using OOPS

```python
class CourseFee:

    def __init__(self, subject, analytics, hostel, food, transport):
        self.subject = subject
        self.analytics = analytics
        self.hostel = hostel
        self.food = food
        self.transport = transport
        self.base_fee = 200000
        self.total = 0

    def calculate_fee(self):

        L1 = ["HR", "Finance", "Marketing", "DS"]

        if self.subject not in L1:
            print("Invalid Subject")
            return

        self.total = self.base_fee

        # Analytics
        if self.analytics == "Y":
            if self.subject != "DS":
                self.total += self.base_fee * 0.10
            else:
                print("Analytics not available for DS")

        # Hostel
        if self.hostel == "Y":
            self.total += 200000

        # Food
        self.total += 2000 * self.food

        # Transport
        if self.transport == "semester":
            self.total += 13000
        elif self.transport == "annual":
            self.total += 13000 * 2

    def print_bill(self):

        print("\n====================================")
        print("          COURSE FEE BILL")
        print("====================================")
        print("Base Course Fee:", self.base_fee)

        if self.analytics == "Y" and self.subject != "DS":
            print("Analytics Fee:", self.base_fee * 0.10)

        if self.hostel == "Y":
            print("Hostel Fee:", 200000)

        print("Food Fee:", 2000 * self.food)

        if self.transport == "semester":
            print("Transport Fee:", 13000)
        elif self.transport == "annual":
            print("Transport Fee:", 13000 * 2)

        print("------------------------------------")
        print("Total Annual Cost =", self.total)
        print("====================================")


subject = input("Enter Subject (HR/Finance/Marketing/DS): ").title()
analytics = input("Analytics (Y/N): ").upper()
hostel = input("Hostel (Y/N): ").upper()
food = int(input("Food months: "))
transport = input("Transport (semester/annual): ").lower()

student = CourseFee(subject, analytics, hostel, food, transport)

student.calculate_fee()
student.print_bill()
```
