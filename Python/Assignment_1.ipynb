11.	Ask user to enter the marks and define if the user got distinction, first class, second class, pass, Fails

If marks are more than 80 output must be “You got distinction”

For marks more than 60 output must be “You got First class”

Marks more than 40 will display “You got second class”

Marks more than or equal to 35 “PASS” Otherwise Fail


```python
#using if-else:
marks=input("Enter a marks:")
if (not marks.isalnum() and not marks.isalpha()) or marks.isdigit():
  marks=int(marks)
  if marks < 0 or marks > 100:
    print("Invalid marks")
  elif marks > 80 and marks <=100:
    print("You got distinction")
  elif marks > 60:
    print("You got First class")
  elif marks > 40 :
    print("You got second class")
  elif marks >=35:
    print("Pass")
  else:
    print("Fail")

else:
  print("Enter a valid marks pls.")
```

#Using the function
1.take input from user that ask which books you want to buy or which notebooks he wants make dictionary of that order

2. calculate overall bill

3.print bill

```python
Books = {
    "Hindi": {'1-4':60,'5-8':100,'9-10':150},
    "Marathi": {'1-4':60,'5-8':100,'9-10':150},
    "English": {'1-4':80,'5-8':100,'9-10':150},
    "Science": {'1-4':90,'5-8':120,'9-10':200},
    "Maths": {'1-4':100,'5-8':140,'9-10':250}
}

Notebooks = {
    "square":{100:40,200:70},
    "4lines":{100:30,200:50},
    "2lines":{100:30,200:50},
    "single lines":{100:60,200:100},
    "a4":{100:100,200:180}
}

receipt = {}
total_cost = 0

print("\n===== Welcome to Book Store =====")

while True:

    choice = input("\nWhat do you want to buy? (Book/Notebook): ").lower()

    # ---------------- BOOK ----------------
    if choice == "book":
        standard = input("Enter class (1-4/5-8/9-10): ")
        subject = input("Enter subject: ").capitalize()
        qty = int(input("Enter quantity: "))

        if subject in Books and standard in Books[subject]:

            unit_price = Books[subject][standard]
            item_total = unit_price * qty
            item_key = subject + " Book"

            total_cost += item_total

            receipt[item_key] = {
                "standard": standard,
                "qty": qty,
                "unit_price": unit_price,
                "total": item_total
            }

        else:
            print("Invalid Book Details!")

    # ---------------- NOTEBOOK ----------------
    elif choice == "notebook":
        note_type = input("Enter notebook type: ").lower()
        pages = int(input("Enter pages (100/200): "))
        qty = int(input("Enter quantity: "))

        if note_type in Notebooks and pages in Notebooks[note_type]:

            unit_price = Notebooks[note_type][pages]
            item_total = unit_price * qty
            item_key = note_type.capitalize() + " Notebook"

            total_cost += item_total

            receipt[item_key] = {
                "pages": pages,
                "qty": qty,
                "unit_price": unit_price,
                "total": item_total
            }

        else:
            print("Invalid Notebook Details!")

    else:
        print("Invalid Choice!")

    more = input("\nDo you want anything more? (Y/N): ").upper()
    if more != 'Y':
        break


# ----------- SHOW STORED DICTIONARY -----------
print("\nStored Receipt Dictionary:\n")
print(receipt)


# ---------------- PRINT RECEIPT ----------------
print("-"*85)
print("|{:^83}|".format("WELCOME TO BOOK STORE"))
print("|{:^83}|".format("RECEIPT"))
print("-"*85)

print("|{:^4}|{:^25}|{:^12}|{:^8}|{:^15}|".format(
    "Sr","Item","Std/Page","Qty","Total"))
print("-"*85)

for index, (item, details) in enumerate(receipt.items(), start=1):

    if "standard" in details:
        extra = details["standard"]
    else:
        extra = str(details["pages"])

    print("|{:^4}|{:^25}|{:^12}|{:^8}|{:^15.2f}|".format(
        index,
        item,
        extra,
        details["qty"],
        details["total"]
    ))

print("-"*85)
print("|{:>65} {:>15.2f} |".format("Grand Total", total_cost))
print("-"*85)

print("\nThank You for Shopping! ")
```

15.	Create an interest bucket for Fix Deposit in the bank. Ask user to enter start date and end date for the FD and check which bucket list it belongs to and assign the interest rate

```python
#simple way
import sys
from datetime import datetime
try:
    start = input("Enter start date (YYYY-MM-DD): ")
    end = input("Enter end date (YYYY-MM-DD): ")

    d1 = datetime.strptime(start, "%Y-%m-%d").date() #converting str to date
    d2 = datetime.strptime(end, "%Y-%m-%d").date()

    days = (d2 - d1).days

    print("Total days:", days)

    if days < 7:
        print("FD should be minimum 7 days")

    else:
        person = input("Enter type (public/senior): ").lower()

        if 7 <= days <= 45:
            if person == "public":
                print("Interest rate is 5.75%")
            elif person == "senior":
                print("Interest rate is 6.25%")

        elif 46 <= days <= 179:
            if person == "public":
                print("Interest rate is 6.25%")
            elif person == "senior":
                print("Interest rate is 6.75%")

        elif 180 <= days <= 210:
            if person == "public":
                print("Interest rate is 6.35%")
            elif person == "senior":
                print("Interest rate is 6.85%")

        elif 211 <= days < 365:
            if person == "public":
                print("Interest rate is 6.40%")
            elif person == "senior":
                print("Interest rate is 6.90%")

        elif 365 <= days < 730:
            if person == "public":
                print("Interest rate is 7.00%")
            elif person == "senior":
                print("Interest rate is 7.50%")

        elif 730 <= days < 1095:
            if person == "public":
                print("Interest rate is 6.75%")
            elif person == "senior":
                print("Interest rate is 7.25%")

        elif 1095 <= days < 1825:
            if person == "public":
                print("Interest rate is 6.70%")
            elif person == "senior":
                print("Interest rate is 7.20%")

        elif 1825 <= days <= 3650:
            if person == "public":
                print("Interest rate is 6.60%")
            elif person == "senior":
                print("Interest rate is 7.10%")

        else:
            print("Duration not available")

except:
  print(sys.exc_info())
```

```python
#using dictionary
from datetime import datetime

# Take input
start = input("Enter start date (YYYY-MM-DD): ")
end = input("Enter end date (YYYY-MM-DD): ")

d1 = datetime.strptime(start, "%Y-%m-%d").date()
d2 = datetime.strptime(end, "%Y-%m-%d").date()

days = (d2 - d1).days

if days < 7:
    print("FD must be minimum 7 days")
else:
    customer = input("Enter customer type (public/senior): ").lower()
    rate_type = input("Enter rate type (existing/revised): ").lower()

    # Interest bucket data
    buckets = [
        (7, 45, 5.75, 5.75, 6.25, 6.25),
        (46, 179, 6.25, 6.25, 6.75, 6.75),
        (180, 210, 6.35, 6.35, 6.85, 6.85),
        (211, 364, 6.40, 6.40, 6.90, 6.90),
        (365, 729, 6.80, 7.00, 7.30, 7.50),
        (730, 1094, 6.80, 6.75, 7.30, 7.25),
        (1095, 1824, 6.80, 6.70, 7.30, 7.20),
        (1825, 3650, 6.85, 6.60, 7.35, 7.10)
    ]

    for bucket in buckets:
        min_d, max_d, pub_old, pub_new, sen_old, sen_new = bucket

        if min_d <= days <= max_d:

            if customer == "public":
                rate = pub_old if rate_type == "existing" else pub_new
            elif customer == "senior":
                rate = sen_old if rate_type == "existing" else sen_new
            else:
                print("Invalid customer type")
                break

            print("Interest Rate:", rate)
            break
    else:
        print("Duration not supported")
```

44.	Create a table  cust_info as sr_no, name, DOB, mobile. Ask user to enter the information from python code. Validate all fields and after validation insert records in the table.

```python
import sqlite3
from datetime import datetime

conn = sqlite3.connect("customer.db")  # Connect to database
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS cust_info (
    sr_no INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    DOB TEXT NOT NULL,
    mobile TEXT NOT NULL
)
""")                              # Creating table

# Name validation
while True:
    name = input("Enter name: ")
    if name.isalpha():
        break
    else:
        print("Invalid name. Only letters allowed.")

# DOB validation
while True:
    dob = input("Enter DOB (YYYY-MM-DD): ")
    try:
        datetime.strptime(dob, "%Y-%m-%d")
        break
    except:
        print("Invalid date format.")

# Mobile validation
while True:
    mobile = input("Enter mobile (10 digits): ")
    if mobile.isdigit() and len(mobile) == 10:
        break
    else:
        print("Invalid mobile number.")

cursor.execute(
    "INSERT INTO cust_info (name, DOB, mobile) VALUES (?, ?, ?)",
    (name, dob, mobile)
)

conn.commit()
conn.close()

print("Record inserted successfully!")
```

```python
#46.
from datetime import date, datetime
def calculate_age(dob_string):
    today = date.today()
    dob = datetime.strptime(dob_string, "%Y-%m-%d").date()

    age = today.year - dob.year

    if (today.month, today.day) < (dob.month, dob.day): # To Check if birthday has not come yet this year
        age -= 1

    return age

dob = "2002-02-14"
print("Age is:", calculate_age(dob))
```

47.	Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.

```python
from datetime import date, datetime
def check_eligibility(dob_string, eligibility):
    today = date.today()
    dob = datetime.strptime(dob_string, "%Y-%m-%d").date()

    age = today.year - dob.year

    if (today.month, today.day) < (dob.month, dob.day):
        age -= 1
    print("Age is:", age)

    if age >= eligibility:
        print("You are eligible")
        return True
    else:
        print("You are not eligible")
        return False

eligibility = 18
dob = "2002-02-14"

check_eligibility(dob, eligibility)
```

54.	Create 3 tables  in mysql having the columns as mentioned
1) user : uid, mobile, passkey, date
2) Product : pid, pname
3) Order : oid, uid, pid

user having  attribute uid, mobile
	      method user sign up
			sign in
			view_products
			view_orders
			place_order
Based on the OOPs concept create classes, methods, attributes and insert information


```python
class Product:

    products = []

    def __init__(self, pid, pname):
        self.pid = pid
        self.pname = pname
        Product.products.append(self)


class Order:

    orders = []

    def __init__(self, oid, uid, pid):
        self.oid = oid
        self.uid = uid
        self.pid = pid
        Order.orders.append(self)


class User:

    users = []

    def __init__(self, uid, mobile, passkey, date):
        self.uid = uid
        self.mobile = mobile
        self.passkey = passkey
        self.date = date
        User.users.append(self)

    def sign_in(self, mobile, passkey):
        if self.mobile == mobile and self.passkey == passkey:
            print("Login successful")
            return True
        else:
            print("Invalid credentials")
            return False

    def view_products(self):
        print("Available Products:")
        for product in Product.products:
            print(product.pid, product.pname)

    def place_order(self, pid):
        oid = len(Order.orders) + 1
        Order(oid, self.uid, pid)
        print("Order placed successfully")

    def view_orders(self):
        print("Your Orders:")
        for order in Order.orders:
            if order.uid == self.uid:
                print("Order ID:", order.oid, "Product ID:", order.pid)


```
